```python
import pandas as pd
import numpy as np
import Recommenders as Recommenders
```

# Importing the data

```python

song_df_1 = pd.read_csv('triplets_file.csv')
song_df_1.head()
```

```python
song_df_2 = pd.read_csv('song_data.csv')
song_df_2.head()
```

# Processing the data

```python

#combine the two datas into a single dataframe 
song_df = pd.merge(song_df_1,song_df_2.drop_duplicates(['song_id']),on ='song_id',how='left')
song_df.head()
```

```python
#check length of the dataframe
len(song_df)
```

```python
print (len(song_df_1),len(song_df_2))
```

```python
# create keyword combining title and artist name 
song_df['song'] = song_df['title']+'-'+song_df['artist_name']
song_df.head()
```

```python
#cumulative sum of listen count - total number of times the song was played
song_grouped = song_df.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
```

```python
# to find percentage of listen count
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage'] = (song_grouped['listen_count']/grouped_sum) * 100
song_grouped.sort_values(['listen_count','song'],ascending=[0,1])

# you can see the most popular songs 
```

#  Popularity Recommendation

```python
ir = Recommenders.item_similarity_recommender_py()
ir.create(song_df,'user_id','song')
```

```python
#past history of the user
user_items = ir.get_user_items(song_df['user_id'][200])
```

```python
for user_item in user_items:
    print(user_item)
```

```python
# song recommendation for the user
ir.recommend(song_df['user_id'][200])
```
